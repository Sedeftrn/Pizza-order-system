```python
import datetime
import csv
#Menu .txt dosyası
from IPython.core.debugger import linecache
dosya = open("Menu.txt", "w") 
dosya.write("Klasik: 5\n")
dosya.write("Margarita: 4\n")
dosya.write("TurkPizza: 7\n")
dosya.write("SadePizza: 3\n")
dosya.write("Mantarli: 5\n")
dosya.write("Sucuklu: 6\n")
dosya.write("Tavuklu: 7\n")
dosya.write("Eksta malzeme ister misiniz?: \n")
dosya.write("Zeytin: 1\n")
dosya.write("Sucuk: 3\n")
dosya.write("Mantar: 2\n")
dosya.write("Peynir: 5\n")
dosya.write("Soğan: 2\n")
dosya.write("Mısır: 1\n")
dosya.close()
#Pizza sınıfının özellikleri ve fiyatları 
class Pizza:
    def __init__(self):
        self.__Klasik = ["Klasik, 5$, Biber, Mantar, Sosis, Zeytin, Mısır, "]
        self.__Margarita = ["Margarita, 4$, Domates, Mozarella, Fesleğen, "]
        self.__TurkPizza = ["TürkPizza, 7$, Domates, Pastırma, Sucuk, Biber, Zeytin, "]
        self.__SadePizza = ["SadePizza, 3$, Biber, Zeytin, Mısır, "]
        self.__Mantarli = ["Mantarli, 5$, Mantar, Biber, Mısır, Kekik, "]
        self.__Sucuklu = ["Sucuklu, 6$, Sucuk, Zeytin, Mısır, Kekik, "]
        self.__Tavuklu = ["Tavuklu, 7$, Tavuk, Biber, Domates, "]

    def add_Klasik(self, item):
        self.__Klasik.append(item) 
    def add_Margarita(self, item):
        self.__Margarita.append(item) 
    def add_TurkPizza(self, item):
        self.__TurkPizza.append(item) 
    def add_SadePizza(self, item):
        self.__SadePizza.append(item) 
    def add_Mantarli(self, item):
        self.__Mantarli.append(item) 
    def add_Sucuklu(self, item):
        self.__Sucuklu.append(item) 
    def add_Tavuklu(self, item):
        self.__Tavuklu.append(item)

    def get_Klasik(self):
        return self.__Klasik
    def get_Margarita(self):
        return self.__Margarita
    def get_TurkPizza(self):
        return self.__TurkPizza
    def get_SadePizza(self):
        return self.__SadePizza
    def get_Mantarli(self):
        return self.__Mantarli
    def get_Sucuklu(self):
        return self.__Sucuklu
    def get_Tavuklu(self):
        return self.__Tavuklu
#Malzeme sınıfının özellikleri ve fiyatları   
class Malzemeler:
    def __init__(self):

      self.__Zeytin = ["1$"]
      self.__Sucuk = ["3$"]
      self.__Mantar = ["2$"]
      self.__Mısır = ["1$"]
      self.__Soğan = ["2$"]
      self.__Peynir = ["5$"]
#Main fonksiyon
def Main():
  siparisler=[]
  malzemeler=[]
  print("Hoşgeldiniz")
  obj = Malzemeler()
  obj = Pizza()
  #Menu .txt dosyasının içindeki verileri sözlük formuna çevirerek ekrana yazdırmak işlemini yapmak için
  with open('Menu.txt', 'r') as dosya:
    sozluk = {}
    for satir in dosya:
      anahtar, deger = satir.strip().split(":")
      sozluk[anahtar] = deger
  for anahtar, deger in sozluk.items():
    print(anahtar, ":", deger)
    price=0
  icerik= input("Pizzaların içeriklerini görmek ister misiniz? (E/H) ")
  if icerik == "E":
    print(obj.get_Klasik())
    print(obj.get_Margarita()) 
    print(obj.get_TurkPizza())
    print(obj.get_SadePizza())
    print(obj.get_Mantarli())
    print(obj.get_Sucuklu())
    print(obj.get_Tavuklu())
#Siparişin kişinin isteğine göre düzenlenrek alınması
  while True:
    pizza = input("İstediğiniz pizza çeşidini giriniz? ")
    if pizza in sozluk:
      price += int(sozluk.get(pizza))
      ekstra = input("Ekstra malzeme ister misiniz?(Evet/Hayır) ")

      if ekstra == "Evet" :
        malzeme = input("Ekstra malzeme giriniz ")
        price += int(sozluk.get(malzeme))

        if pizza=="Klasik":
           obj.add_Klasik(malzeme)
           print(obj.get_Klasik())

        elif pizza=="Margarita":
           obj.add_Margarita(malzeme)
           print(obj.get_Margarita())

        elif pizza=="TurkPizza":
           obj.add_TurkPizza(malzeme)
           print(obj.TurkPizza())  

        elif pizza=="SadePizza":
           obj.add_SadePizza(malzeme)
           print(obj.get_SadePizza())

        elif pizza=="Mantarli":
           obj.add_Mantarli(malzeme)
           print(obj.get_Mantarli())

        elif pizza=="Sucuklu":
           obj.add_Sucuklu(malzeme)
           print(obj.get_Sucuklu())

        elif pizza=="Tavuklu":
           obj.add_Tavuklu(malzeme)
           print(obj.get_Tavuklu())
        malzemeler.append(malzeme)
    siparisler.append(pizza)
    secim = input("Seçimiz bitti mi? (E/H) ")
    if secim== 'E':
        break
  #Verdiği siparişin ekrana yazdırılması
  print("Sipariş ettiğiniz pizzalar", siparisler , "İstediğiniz ekstra malzemeler", malzemeler)
  simdi = datetime.datetime.now()
  print("Ödemeniz gereken toplam fiyat =", price,"$")  
  kullanici_adi = input("Lütfen kullanıcı adınızı giriniz: ")
  kimlik_numarası = input("Lütfen TC kimlik numaranızı giriniz: ")
  kart_numarasi = input("Lütfen kredi kart numaranızı giriniz: ")
  kart_sifresi = input("Lütfen kredi kart sifrenizi giriniz: ")
#Verilen siparişin ve alınan verilerin tutulduğu liste
  bilgiler = []
  bilgiler.append(siparisler)
  bilgiler.append(malzemeler)
  bilgiler.append(str(kullanici_adi))
  bilgiler.append(str(kimlik_numarası))
  bilgiler.append(str(kart_numarasi))
  bilgiler.append(str(kart_sifresi))
  saat=[simdi.hour,":",simdi.minute,":",simdi.second]
  bilgiler.append(str(saat))
  print(bilgiler)
  with open("bilgiler.csv", mode="w", newline="") as dosya:
    yazici = csv.writer(dosya)
    yazici.writerows(bilgiler)

Main()
```
